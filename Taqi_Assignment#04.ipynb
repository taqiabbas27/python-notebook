```python
# Question 4

course_a = {"John", "Alice", "Bob", "David"}
course_b = {"Alice", "Eve", "Charlie", "David"}

difference = course_a.difference(course_b)
sym_diff = course_a.symmetric_difference(course_b)

print("Students in Course A but not in Course B:", difference)
print("Students only in one of the two courses:", sym_diff)
```

```python
# Question 7

inventory = {
    "The Great Gatsby": {"price": 10.99, "quantity": 5},
    "Rich dad poor dad": {"price": 8.99, "quantity": 8},
    "Harry Potter": {"price": 12.50, "quantity": 0}
}

inventory.update({
    "Atomic Habits": {"price": 15.99, "quantity": 6}
})

print("Book Titles:", inventory.keys())

if inventory["Harry Potter"]["quantity"] == 0:
    inventory.pop("Harry Potter")

print(inventory)
```
