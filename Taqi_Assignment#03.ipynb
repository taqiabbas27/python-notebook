```python
# Question 7

code = "SKU-12345-XYZ"

position = code.find("-")
count = code.count("-")
number = code.removeprefix("SKU-").removesuffix("-XYZ")

print("First dash position:", position)
print("Number of dashes:", count)
print("Numeric part:", number)
```
